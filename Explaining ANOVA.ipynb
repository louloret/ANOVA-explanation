# Hypothesis Test with F- statistic

```python
# example: does type of food eaten before test affect your score?
        # is difference in mean test scores due to random chance? Or are there 
         # actual differences across food group means
```

#### Ho = mean1 = mean2 = mean3

```python
# null hypothesis is that food does not make a difference
# true population of means will be the same
```

#### F-stat = (SSB/degrees of freedom between)/(SSW/degrees of freedom within)

```python
# a larger number would indicate there is actually a difference between 
# food group test scores (lower probability that null hypothesis is correct)
```

#### Fstat = (24/2)/(6/6)  = 12/1 = 12

```python
# let's say we wanted a sig level of alpha = 0.10
```

```python
# given the f distribution, which defines your f critical value
# based on the degrees of freedom in your numerator and denominator 
```

```python
Image(filename = path + "f_distr.png")
```

```python
# critical f-value in this case is 3.46
```

```python
# our F-stat is much larger than this value, because of that we reject our 
#null hypothesis
# meaning that the probability of seeing this much difference in mean
# is very small, there probably is an effect on what food group you belong to
# and test scores
```
